## Step 3: Data Preparation

#### Mapping the columns with 0s and 1s as to convert it to numerical columns

```python
Car_Details.fueltype = Car_Details.fueltype.map({"gas":0,"diesel":1}) # converting fueltype as 0 and 1 
```

```python
Car_Details.aspiration = Car_Details.aspiration.map({"std":0,"turbo":1}) # converting aspiration as 0 and 1 
```

```python
Car_Details.doornumber = Car_Details.doornumber.map({"two":0,"four":1}) # converting doornumber as 0 and 1 
```

```python
Car_Details.enginelocation = Car_Details.enginelocation.map({"front":0,"rear":1}) # converting enginelocation as 0 and 1 
```

#### Creating Dummy variables for categorical variables whose value is more than 2 

```python
carbody = pd.get_dummies(Car_Details["carbody"], drop_first = True)
carbody.head()
```

```python
drivewheel = pd.get_dummies(Car_Details["drivewheel"], drop_first = True)
drivewheel.head(3)
```

```python
enginetype = pd.get_dummies(Car_Details["enginetype"], drop_first = True)
enginetype.head(3)
```

```python
cylinder = pd.get_dummies(Car_Details["cylindernumber"], drop_first = True)
cylinder.head(3)
```

```python
fuelsys = pd.get_dummies(Car_Details["fuelsystem"], drop_first = True)
fuelsys.head(3)
```

```python
# Dropping categorical columns which have more than 2 values
Car_Details.drop(['carbody','fuelsystem','cylindernumber','enginetype','drivewheel','car_ID'],axis=1,inplace=True)
```

```python
# Concating the dataset with dummy columns
Car_Details = pd.concat([Car_Details,carbody,fuelsys,cylinder,enginetype,drivewheel],axis=1)
Car_Details.columns
```

```python
Car_Details.shape # looking at the shape of dataset onces its concated with original dataset
```

```python
# Separating the modelname and company name from the carname column from the dataset
Car_Details[['companyname','modelname']] = Car_Details.CarName.str.split(' ',1,expand=True)
```

```python
# dropping Columns carname and modelname 
Car_Details.drop(['modelname','CarName'],axis=1,inplace=True)
```

```python
Car_Details.head(3)
```

```python
# correcting the data with proper name if the company

Car_Details["companyname"]= Car_Details.companyname.replace(to_replace =["Nissan","vokswagen","vw","toyouta","maxda","porcshce"],
                                                         value =["nissan","volkswagen","volkswagen","toyota","mazda","porsche"])
Car_Details.companyname.unique()
```

```python
# Creating dummies for Company name column as it is a categorical column and it should be converted to numeric to numeric to perform linear regression
CompanyName = pd.get_dummies(Car_Details["companyname"], drop_first = True)
CompanyName.head(3)
```

```python
# Concatinating the Car_details with companyname column
Car_Details = pd.concat([Car_Details,CompanyName],axis=1)
```

```python
Car_Details.shape # looking at the shape after concatination
```

```python
Car_Details.drop(["companyname"],axis=1,inplace=True) # Droping the column company name from the dataset as it is no longer required
```
